```python
############# Packages #############
import numpy as np 
import matplotlib.pyplot as plt 
########### plt ##########
plt.rcParams["font.family"] = "serif"       # fonts
plt.rcParams["font.serif"] = "Times New Roman"
plt.rcParams["font.size"] = 18              
plt.rcParams["mathtext.cal"] = "serif"      
plt.rcParams["mathtext.rm"] = "serif"       
plt.rcParams["mathtext.it"] = "serif:italic"
plt.rcParams["mathtext.bf"] = "serif:bold"  
plt.rcParams["mathtext.fontset"] = "cm"     
#################################
```

```python
############# dynamics #############
def dynamics_transcritical(mu, x):
    return x*(mu - x)

############# plot #############
def plot_dynamics(x):
    mu_lst = [-1.0, 0.0, 1.0]
    dyn_mu = []
    for mu in mu_lst:
        dyn_mu.append(dynamics_transcritical(mu=mu, x=x))
    fig, axes = plt.subplots(1, 3, figsize=(16,6))
    plt.subplots_adjust(wspace=0.4, hspace=0.6)
    for i in range(len(mu_lst)):
        axes[i].plot(x, dyn_mu[i], c="k")
        axes[i].set_xlim([x[0], x[-1]])
        axes[i].set_xlabel(r"$x$"); axes[i].set_ylabel(r"$f(x, \mu)$")
        #axes[i].grid(True)
        axes[i].text(0.05, 0.3, r"$\mu = {}$".format(mu_lst[i]))
        axes[i].set_ylim([-1, 0.5])
        xl = [-2.5, 2.5]; yl = [0, 0]
        xs = [0, 0]; ys = [-2, 1]
        axes[i].plot(xl, yl, c='k', linewidth=1.25)     # x軸
        axes[i].plot(xs, ys, c='k', linewidth=1.25)     # y軸
    axes[0].annotate(text='',xy=(0,0), xytext=(2.5,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    axes[0].annotate(text='',xy=(0,0), xytext=(-1,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    axes[0].annotate(text='',xy=(-2.5,0), xytext=(-1,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    axes[1].annotate(text='',xy=(0,0), xytext=(2.5,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    axes[1].annotate(text='',xy=(-2.5,0), xytext=(0,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    axes[2].annotate(text='',xy=(1,0), xytext=(2.5,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    axes[2].annotate(text='',xy=(1,0), xytext=(0,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    axes[2].annotate(text='',xy=(-2.5,0), xytext=(0,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    plt.savefig("./6.3_transcritical_graph.svg")
```

```python
x = np.linspace(-2.5, 2.5, 401)
plot_dynamics(x)
```
