```python
############# Packages #############
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
############# dynamics #############
def dynamics_transcritical(mu, x):
    return x*(mu - x)

############# plot #############
def plot_dynamics(x):
    mu_lst = [-1.0, 0.0, 1.0]
    dyn_mu = []
    for mu in mu_lst:
        dyn_mu.append(dynamics_transcritical(mu=mu, x=x))
    fig, axes = plt.subplots(1, 3, figsize=(16,6))
    plt.subplots_adjust(wspace=0.4, hspace=0.6)
    for i in range(len(mu_lst)):
        axes[i].plot(x, dyn_mu[i], c="k")
        axes[i].set_xlim([x[0], x[-1]])
        axes[i].set_xlabel(r"$x$"); axes[i].set_ylabel(r"$f(x, \mu)$")
        axes[i].grid(True)
        axes[i].text(0, 0.5, r"$\mu = {}$".format(mu_lst[i]))
        axes[i].set_ylim([-2, 1])
    plt.savefig("./6.3_transcritical_graph.svg")
```

```python
x = np.linspace(-1.5, 1.5, 401)
plot_dynamics(x)
```

```python
def plot_plane():
    mu_array_minus = np.linspace(-1, 0, 201)
    zero_array_minus = np.zeros_like(mu_array_minus)
    mu_array_plus = np.linspace(0, 1, 201)
    zero_array_plus = np.zeros_like(mu_array_plus)
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.plot(mu_array_minus, zero_array_minus, c="k", linestyle="solid")
    ax.plot(mu_array_minus, mu_array_minus, c="k", linestyle=(0, (3, 7.5)))
    ax.plot(mu_array_plus, zero_array_plus, c="k", linestyle=(0, (3, 7.5)))
    ax.plot(mu_array_plus, mu_array_plus, c="k", linestyle="solid")
    ax.set_xlim([-1,1])
    ax.set_xlabel(r"$\mu$"); ax.set_ylabel(r"$x$")
    plt.savefig("./6.3_mu_x_plane.svg")
```

```python
plot_plane()
```
