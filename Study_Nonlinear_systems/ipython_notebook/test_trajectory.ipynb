```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import solve_ivp
```

```python
def dynamics(t, y):
    return -y**2 

def solve_dynamics(yinit, T, N):
    sol = solve_ivp(fun=dynamics, t_span=[0,T], y0=yinit, dense_output=True)
    t = np.linspace(0, T, N)
    z = sol.sol(t)
    return z 
```

```python
T = 200; N = 201
yinit = np.array([0.25])
sol_plus = solve_dynamics(yinit, T, N)
t_plus = np.linspace(0, T, N)

T = 2.5; N = 51
yinit = np.array([-0.25])
sol_minus = solve_dynamics(yinit, T, N)
t_minus = np.linspace(0, T, N)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(t_plus, sol_plus[0])
ax.set_xlabel(r'$t$'); ax.set_ylabel(r'$x$')
plt.show()
```

```python
plt.cla()
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(t_minus, sol_minus[0])
ax.set_xlabel(r'$t$'); ax.set_ylabel(r'$x$')
plt.show()
```
