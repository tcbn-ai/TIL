```python
############# Packages #############
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def dynamics_saddle(mu, x, y):
    dx = mu - x**2
    dy = (-1)*y
    return dx, dy 
```

```python
X, Y = np.meshgrid(np.linspace(-2, 2, 41), np.linspace(-2, 2, 41))
```

```python
def plot_quiver(x, y):
    mu_lst = [-1, 0, 1]
    fig, axes = plt.subplots(1, 3, figsize=(16,6))
    plt.subplots_adjust(wspace=0.4, hspace=0.6)
    for i in range(len(mu_lst)):
        dX, dY = dynamics_saddle(mu_lst[i], X, Y)
        axes[i].quiver(X, Y, dX, dY)
        axes[i].set_xlabel("x"); axes[i].set_ylabel("y")
        axes[i].set_title(r"$\mu = {}$".format(mu_lst[i]))
    plt.savefig("./6.5_saddle_node_2dim_vector_quiver.svg")
```

```python
plot_quiver(X, Y)
```
