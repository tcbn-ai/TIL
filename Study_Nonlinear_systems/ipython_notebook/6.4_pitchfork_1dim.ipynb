```python
############# Packages #############
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def dynamics_pitchfork(mu, x):
    return x*(mu - x**2)

def plot_dynamics(x):
    mu_lst = [-1.0, 0.0, 1.0]
    dyn_mu = []
    for mu in mu_lst:
        dyn_mu.append(dynamics_pitchfork(mu, x))
    fig, axes = plt.subplots(1, 3, figsize=(16,6))
    plt.subplots_adjust(wspace=0.4, hspace=0.6)
    for i in range(len(mu_lst)):
        axes[i].plot(x, dyn_mu[i], c="k")
        axes[i].set_xlim([x[0], x[-1]])
        axes[i].set_xlabel(r"$x$"); axes[i].set_ylabel(r"$f(x, \mu)$")
        # axes[i].grid(True)
        axes[i].set_ylim([-2.0, 2.0])
        axes[i].text(0.05, 0.5, r"$\mu = {}$".format(mu_lst[i]))
        xl = [-1.5, 1.5]; yl = [0, 0]
        xs = [0, 0]; ys = [-2, 2]
        axes[i].plot(xl, yl, c='k', linewidth=1.25)     # x軸
        axes[i].plot(xs, ys, c='k', linewidth=1.25)     # y軸
    axes[0].annotate(text='',xy=(0,0), xytext=(1.5,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    axes[0].annotate(text='',xy=(0,0), xytext=(-1.5,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    axes[1].annotate(text='',xy=(0,0), xytext=(1.5,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    axes[1].annotate(text='',xy=(0,0), xytext=(-1.5,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    axes[2].annotate(text='',xy=(1,0), xytext=(1.5,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    axes[2].annotate(text='',xy=(1,0), xytext=(0,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    axes[2].annotate(text='',xy=(-1,0), xytext=(0,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    axes[2].annotate(text='',xy=(-1,0), xytext=(-1.5,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    plt.savefig("./6.4_pitchfork_graph.svg")
```

```python
x = np.linspace(-1.5, 1.5, 201)
plot_dynamics(x)
```

```python
def P_plus(mu):
    return np.sqrt(mu)

def P_minus(mu):
    return (-1) * np.sqrt(mu)

def plot_plane():
    mu_array_minus = np.linspace(-1, 0, 201)
    zero_array_minus = np.zeros_like(mu_array_minus)
    mu_array_plus = np.linspace(0, 1, 201)
    zero_array_plus = np.zeros_like(mu_array_plus)
    P_plus_array = P_plus(mu_array_plus)
    P_minus_array = P_minus(mu_array_plus)
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.plot(mu_array_minus, zero_array_minus, c="k", linestyle="solid")
    ax.plot(mu_array_plus, zero_array_plus, c="k", linestyle=(0, (3, 7.5)))
    ax.plot(mu_array_plus, P_plus_array, c="k", linestyle="solid")
    ax.plot(mu_array_plus, P_minus_array, c="k", linestyle="solid")
    ax.set_xlim([-1,1]); ax.set_ylim([-1, 1])
    xs = [0, 0]; xs2 = [0.5, 0.5]; xs3 = [-0.5, -0.5]; ys = [-1, 1]
    ax.plot(xs, ys, c='k', linewidth=1.25)
    ax.plot(xs2, ys, c='k', linewidth=1.25)
    ax.plot(xs3, ys, c='k', linewidth=1.25)
    ax.annotate(text='',xy=(0,0), xytext=(0,1), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    ax.annotate(text='',xy=(0,0), xytext=(0,-1), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    ax.annotate(text='',xy=(0.5,np.sqrt(0.5)), xytext=(0.5,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    ax.annotate(text='',xy=(0.5,np.sqrt(0.5)), xytext=(0.5,1), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    ax.annotate(text='',xy=(0.5,-np.sqrt(0.5)), xytext=(0.5,0), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    ax.annotate(text='',xy=(0.5,-np.sqrt(0.5)), xytext=(0.5,-1), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    ax.annotate(text='',xy=(-0.5,0), xytext=(-0.5,1), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    ax.annotate(text='',xy=(-0.5,0), xytext=(-0.5,-1), xycoords='data',\
            arrowprops=dict(facecolor='black', width=0, headwidth=7.0, headlength=7.0, shrink=0.4))
    ax.set_xlabel(r"$\mu$"); ax.set_ylabel(r"$x$")
    plt.savefig("./6.4_mu_x_plane.svg")
```

```python
plot_plane()
```
